```python
#imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.lines as mlines

#loading data
df = pd.read_csv('health.csv')
df.head(3)

#sorting data and reseting index
df.sort_values('pct_2014', inplace=True)
df.reset_index(inplace=True)

#function to draw line segment
def newline(p1,p2,color='black'):
    ax = plt.gca()
    l = mlines.Line2D([p1[0],p2[0]], [p1[1],p2[1]], color='skyblue')
    ax.add_line(l)
    return l 

#defining figure & axes
fig, ax = plt.subplots(1,1,figsize=(14,14), dpi=80, facecolor='#f7f7f7')

#Vertical Lines
ax.vlines(x=.05, ymin=0, ymax=26,color='black', alpha=0.7, linewidth=1, linestyles='dotted')
ax.vlines(x=.10, ymin=0, ymax=26, color='black', alpha=0.7, linewidth=1, linestyles='dotted')
ax.vlines(x=.15, ymin=0, ymax=26, color='black', alpha=0.7, linewidth=1, linestyles='dotted')
ax.vlines(x=.20, ymin=0, ymax=26, color='black', alpha=0.7, linewidth=1, linestyles='dotted')

#Scattered points
ax.scatter(y=df.index, x=df['pct_2013'], s=50, color='#0e668b', alpha=0.7)
ax.scatter(y=df.index, x=df['pct_2014'], s=50, color='#a3c4dc', alpha=0.7)

#Line Segments
for i,p1,p2 in zip(df.index, df['pct_2013'],df['pct_2014']):
    newline([p1,i],[p2,i])
    
#Decorations
ax.set_facecolor('#f7f7f7')
ax.set_title('Dumbell Chart: Pct Change - 2013 vs 2014', fontdict={'size':22})
ax.set(xlim=(0,.25), ylim=(-1,27), ylabel='Mean GDP Per Capita')
ax.set_xticks([0.05,0.15,0.20, 0.25])
ax.set_xticklabels(['5%','15%','20%','25%'])
plt.show()
```
