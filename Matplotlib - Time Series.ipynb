```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#load Dataset
df = pd.read_csv('AirPassengers.csv')

#prepare dataset 
datelist = df['date'].tolist()
trafficlist = df['traffic'].tolist()

#Draw Plor
plt.figure(figsize=(16,10), dpi=80)
plt.plot(datelist,trafficlist, color='tab:red' )

#Decoration
plt.title('Air Passengers Traffic (1949 - 1969)', fontsize=22)
plt.yticks(fontsize=16, alpha=0.3)
plt.ylim(50,750)

xtick_location = df.index.tolist()[::12]
xtick_label = [x[-4:] for x in df.date.tolist()[::12]]
plt.xticks(ticks=xtick_location, labels=xtick_label, fontsize=16, alpha=0.7, horizontalalignment='center', rotation=60)
plt.grid(axis='both', alpha=0.7)
```
