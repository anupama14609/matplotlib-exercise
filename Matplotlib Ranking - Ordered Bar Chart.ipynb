```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import warnings; warnings.simplefilter('ignore')
df_raw = pd.read_csv('mpg_ggplot3.csv')
df = df_raw[['cty','manufacturer']].groupby('manufacturer').apply(lambda x:x.mean())
df.sort_values('cty', inplace=True)
df.reset_index(inplace=True)

fig, ax = plt.subplots(figsize=(16,10), dpi=80, facecolor='white')
ax.vlines(df.index, ymin=0, ymax=df.cty, color='firebrick', alpha=.7, linewidth=20)

#Annotate
for i, cty in enumerate(df.cty):
    ax.text(i,cty+0.5, round(cty,1), horizontalalignment='center')
    
# Title, Label, Ticks and Ylim
ax.set_title('Bar Chart for Highway Mileage', fontdict={'size':22})
ax.set(ylabel='Miles Per Gallon', ylim=(0,30))
plt.xticks(df.index, df.manufacturer.str.upper(), rotation=60, horizontalalignment='right', fontsize=12)

# Add patches to color the X axis labels
p1 = patches.Rectangle((.57, -0.005), width=.33, height=.13, alpha=0.1, facecolor='green', transform=fig.transFigure)
p2 = patches.Rectangle((.124, -0.005), width=.446, height=.13, alpha=0.1, facecolor='red', transform=fig.transFigure)

fig.add_artist(p1)
fig.add_artist(p2)

plt.show()
```
