```python
import pandas as pd
import warnings; warnings.simplefilter('ignore')
import matplotlib.pyplot as plt 

df_raw = pd.read_csv('mpg_ggplot3.csv')
df = df_raw[['manufacturer', 'cty']].groupby('manufacturer').apply(lambda x: x.mean())
df.sort_values('cty', inplace=True)
df.reset_index(inplace=True)

#Draw Plot
fig, ax = plt.subplots(figsize=(16,10), dpi=80)
ax.vlines(x= df.index, ymin=0, ymax=df.cty, color='red',alpha=0.7, linewidth=2)
ax.scatter(x=df.index, y=df.cty, s=75, alpha=0.7, color='firebrick')

## Title, Label, Ticks and Ylim
ax.set_title('Lollipop Chart for Highway Mileage', fontdict={'size':22})
ax.set_ylabel('Miles Per Gallon')
ax.set_xticks(df.index)
ax.set_xticklabels(df.manufacturer.str.upper(), rotation=60, fontdict={'size':12,'horizontalalignment':'right'})
ax.set_ylim(0, 30)

#Annotate
for row in df.itertuples():
    ax.text(row.Index, row.cty+.5, s=round(row.cty,2), horizontalalignment='center', verticalalignment='bottom', fontsize=14)
plt.show()
```
